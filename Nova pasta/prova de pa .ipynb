```python
hp = int(input("Digite a energia do seu personagem: "))

print(f"{hp} de energia")


if (hp >= 71) and (hp <= 100):
    print(f"O seu personagem ainda está com {hp} de energia, ele está saudável")

if (hp >= 31) and (hp <= 70):
    print(f"O seu personagem ainda está com {hp} de energia, ele está Normal")

if (hp >= 1) and (hp <= 30):
    print(f"O seu personagem ainda está com {hp} de energia, ele está em estado Critico")

if hp == 0:
    print(f"Você está com {hp} de energia, você está Morto")
```
