```python
# Operadores aritiméticos


# Soma de dois numeros

# Entrada
a =int(input("digite o primeiro numero: "))
b =int(input("digite o segundo numero: "))

# Processamento

# Soma  de dois numeros
soma = a + b
subtraçao = a - b
mutiplicaçao = a * b
divisao = a / b
div_int = a//b


# Saida
print("")
print("------------soma-----------")
print(f"a soma de {a} + {b} = {soma}")
print("")
print("------------SUBTRAÇÃO---------")
print(f"a subtração de {a} - {b} = {subtraçao}")
print("")
print("------------MULTIPLICAÇÃO-------")
print(f"a Mutipicação de {a} * {b} = {mutiplicaçao}")
print("")
print("------------DIVISÃO--------")
print(f"a divisão de {a} / {b} = {divisao}")
print("")
print("------------DIVISÃO INTEIRA--------")
print(f"a divisão inteira de {a} // {b} = {div_int}")
print("---------------------------")
```
